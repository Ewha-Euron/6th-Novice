
PyCaret은 Python에서 사용할 수 있는 머신 러닝 라이브러리 중 하나로, 데이터 사이언스를 위한 오픈 소스, 저코드(low-code) 라이브러리입니다. 이 라이브러리는 데이터 전처리, 모델 학습, 모델 평가, 모델 튜닝 등 데이터 사이언스 프로젝트의 전반적인 프로세스를 간소화하고 자동화하는 데 초점을 맞추고 있습니다. 사용자가 복잡한 코드를 작성하지 않고도 머신 러닝 모델을 쉽게 만들고 평가할 수 있도록 설계되었습니다.afe
MarkupSafe는 문자열을 안전하게 처리하는 데 사용되는 Python 라이브러리입니다. 주로 HTML과 XML에서 사용되는 특수 문자를 이스케이프 처리하여 XSS(cross-site scripting) 공격과 같은 보안 문제를 방지하는 데 사용됩니다. Jinja2와 같은 템플릿 엔진에서 HTML 코드를 동적으로 생성할 때 MarkupSafe를 사용하여 안전하게 문자열을 처리합Jinja2
Jinja2는 Python에서 사용되는 가장 인기 있는 템플릿 엔진 중 하나입니다. Flask와 같은 웹 프레임워크에서 주로 사용되며, HTML 템플릿에 Python 코드를 삽입하여 동적인 웹 페이지를 생성할 수 있습니다. Jinja2는 템플릿 상속, 매크로, 자동 이스케이프 처리 등 다양한 기능을 제공하여 복잡한 웹 애플리케이션의 프론트엔드를 구성하는 데 유용합니다. Jinja2는 MarkupSafe와 함께 작동하여 템플릿에서 생성된 문자열이 안전하게 처리되도록 합니다.

```python
# --- Importing Libraries ---
import datetime
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import scipy
import pycaret
import warnings
import jinja2

from sklearn.preprocessing import LabelEncoder
from scipy import stats
from scipy.stats import *
from pycaret.regression import *

# --- Libraries Settings ---
warnings.filterwarnings('ignore')
sns.set_style('whitegrid')
plt.rcParams['figure.dpi'] = 100
```

```python
# --- Create List of Color Palletes ---
olive_grad = ['#2A3F07', '#557F0F', '#7FBE17', '#A4E637', '#C1EE76']
yellow_grad = ['#413D05', '#837A0B', '#C5B710', '#E8DA2E', '#F9EE6A']
color_mix = ['#356211', '#6C8F32', '#F2E880', '#C15C37', '#AA471F']
black_grad = ['#100C07', '#3E3B39', '#6D6A6A', '#9B9A9C', '#CAC9CD']

# --- Plot Color Palletes --
sns.palplot(olive_grad)
sns.palplot(yellow_grad)
sns.palplot(color_mix)
sns.palplot(black_grad)
```

sns.palplot 함수는 seaborn 라이브러리를 사용하여 주어진 색상 팔레트를 시각화한다.

```python
# --- Reading Dataset ---
ds = pd.read_csv('./avocado.csv')
ds.head().style.background_gradient(cmap='Greens').set_properties(**{'font-family': 'Segoe UI'}).hide()
```

```python
# --- Print Dataset Info ---
print('\033[92m\033[1m'+'.: Imported Dataset Info :.'+'\033[0m')
print('\033[92m*' * 30+'\033[0m')
print('Total Rows:'+'\033[92m\033[1m', ds.shape[0])
print('\033[0m'+'Total Columns:'+'\033[92m\033[1m', ds.shape[1])
print('\033[0m\033[92m'+'*' * 30+'\033[0m')
print('\n')

# --- Print Dataset Detail ---
print('\033[92m\033[1m'+'.: Dataset Details :.'+'\033[0m')
print('\033[0m\033[92m'+'*' * 30+'\033[0m')
ds.info(memory_usage = False)
```

Python의 ANSI 이스케이프 시퀀스를 사용하여 콘솔이나 터미널에 데이터셋 정보를 색상과 함께 출력한다.

```python
# --- Colors and Labels (Null Values Dropped) ---
colors = color_mix[1:3]
labels = ds['type'].dropna().unique()
order = ds['type'].value_counts().index

# --- Size for Both Figures ---
plt.figure(figsize=(18, 10))

# --- Pie Chart ---
plt.subplot(1, 2, 1)
plt.title('Avocado Type Percentage', fontweight = 'bold', fontsize = 14, fontfamily = 'sans-serif', 
          color = black_grad[0])
plt.pie(ds['type'].value_counts(), labels = order, colors = colors, 
        wedgeprops = dict(alpha = 0.8, edgecolor = black_grad[1]), autopct = '%.2f%%')

# --- Bar Chart ---
countplt = plt.subplot(1, 2, 2)
plt.title('Avocado Type Bar Chart', fontweight = 'bold', fontsize = 14, fontfamily = 'sans-serif', color = black_grad[0])
ax = sns.countplot(x = 'type', data = ds, palette = colors, order = order, edgecolor = black_grad[2], alpha = 0.85)
for rect in ax.patches:
    ax.text (rect.get_x()+rect.get_width()/2, rect.get_height()+0.75,rect.get_height(), horizontalalignment = 'center', 
             fontsize = 11)

plt.xlabel('Types', fontweight = 'bold', fontsize = 11, fontfamily = 'sans-serif', color = black_grad[1])
plt.ylabel('Total', fontweight = 'bold', fontsize = 11, fontfamily = 'sans-serif', color = black_grad[1])
plt.grid(axis = 'y', alpha = 0.4)
countplt

# --- Count Categorical Labels w/out Dropping Null Walues ---
print('\033[92m'+'*' * 40)
print('\033[92m\033[1m'+'.: Total Avocado based on each Type :.'+'\033[0m')
print('\033[92m'+'*' * 40+'\033[0m')
ds.type.value_counts(dropna=False)
```

먼저, color_mix에서 특정 색상을 선택하여 colors 배열을 생성하고, ds['type']에서 결측값을 제외한 고유값을 labels로 설정한다. 
아보카도 유형별 빈도수를 기준으로 정렬된 순서는 order에 저장된다.
이후, matplotlib 라이브러리를 사용하여 두 가지 유형의 차트를 그린다.
    
Pie Chart(파이 차트): 아보카도 유형별 비율을 나타낸다. 
plt.pie 함수를 사용하여, 각 유형별 비율, 색상, 테두리 설정, 백분율 표시 등을 설정하고 차트의 제목에는 black_grad 색상 배열에서 색상을 선택하여 적용한다.

Bar Chart(막대 차트): 아보카도 유형별 총량을 나타낸다. 
sns.countplot 함수를 사용하여 데이터셋에서 각 유형별 개수를 막대 그래프로 표현한다. 
막대의 테두리 색상과 투명도를 설정하고, 각 막대 상단에 해당 값이 표시되도록 한다. x축과 y축의 라벨에도 black_grad 색상 배열을 이용해 색상을 적용한다.

마지막으로, ds.type.value_counts(dropna=False)를 사용하여 아보카도 유형별 총 개수를 출력하며, 여기서는 결측값도 포함하여 계산한다.

```python
# --- Colors and Labels (Null Values Dropped) ---
colors = sns.color_palette('YlOrBr')
labels = ds['year'].dropna().unique()
order = ds['year'].value_counts().index

# --- Size for Both Figures ---
plt.figure(figsize=(18,10))

# --- Pie Chart ---
plt.subplot(1,2,1)
plt.title('Avocado Years Percentage', fontweight = 'bold', fontsize = 14, fontfamily = 'sans-serif', color = black_grad[0])
plt.pie(ds['year'].value_counts(), labels = order, colors = colors, 
        wedgeprops = dict(alpha = 0.8, edgecolor = black_grad[1]), autopct = '%.2f%%')

# --- Bar Chart ---
countplt = plt.subplot(1,2,2)
plt.title('Avocado Years Bar Chart', fontweight = 'bold', fontsize = 14, fontfamily = 'sans-serif', color = black_grad[0])
ax = sns.countplot(x = 'year', data = ds, palette = colors, order = order, edgecolor = black_grad[2], alpha = 0.85)
for rect in ax.patches:
    ax.text (rect.get_x()+rect.get_width()/2, rect.get_height()+0.75,rect.get_height(), horizontalalignment = 'center',
             fontsize = 10)

plt.xlabel('Years', fontweight = 'bold', fontsize = '11', fontfamily = 'sans-serif', color = black_grad[1])
plt.ylabel('Total', fontweight = 'bold', fontsize = '11', fontfamily = 'sans-serif', color = black_grad[1])
plt.grid(axis = 'y', alpha = 0.4)
countplt

# --- Count Categorical Labels w/out Dropping Null Walues ---
print('\033[92m'+'*' * 35)
print('\033[92m\033[1m'+'.: Total Avocado based on Year :.'+'\033[0m')
print('\033[92m'+'*' * 35+'\033[0m')
ds.year.value_counts(dropna=False)
```

이 코드는 데이터셋(ds) 안에 있는 아보카도의 연도별 분포를 시각화하기 위해 작성되었다. sns.color_palette('YlOrBr')를 사용하여 색상 팔레트를 설정하고, labels에 연도별 고유값을 추출하여 저장한다. 아보카도 데이터를 연도별로 빈도수에 따라 정렬하여 order에 저장한다.

데이터 시각화는 두 가지 차트로 진행된다:

파이 차트: plt.pie 함수를 사용하여 연도별 아보카도의 비율을 나타낸다. 각 조각의 색상, 테두리 설정 및 비율 표시 방법을 정의한다. 차트 제목에는 black_grad 배열에서 선택한 색상을 적용한다.

막대 차트: sns.countplot 함수를 사용하여 연도별 아보카도의 총 수량을 막대 그래프로 표시한다. 막대의 테두리 색상과 투명도를 설정하고, 각 막대 위에 해당 값이 표시되도록 한다. x축과 y축 레이블에는 black_grad 배열에서 선택한 색상을 적용한다.

마지막으로, 누락된 값까지 포함하여 연도별 아보카도의 총 수량을 출력한다. ANSI 이스케이프 시퀀스를 사용하여 출력 텍스트 스타일을 설정한다.
