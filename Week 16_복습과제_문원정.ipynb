```python
import pandas as pd
import numpy as np
import warnings; warnings.filterwarnings('ignore')

movies =pd.read_csv('./tmdb_5000_movies.csv')
print(movies.shape)
movies.head(1)
```

```python
movies_df = movies[['id','title', 'genres', 'vote_average', 'vote_count',
                 'popularity', 'keywords', 'overview']]
```

'movies' DataFrame에서 특정 열들을 선택하여 새로운 DataFrame 'movies_df'를 만든다.
'movies_df'에는 영화의 ID, 제목, 장르, 평점, 투표 수, 인기, 키워드, 개요 정보가 포함된다

pandas 옵션을 설정하여 DataFrame 출력 시 열의 최대 너비를 100으로 지정
'movies_df' DataFrame에서 'genres'와 'keywords' 열의 첫 번째 행만 선택하여 출력

```python
from ast import literal_eval

movies_df['genres'] = movies_df['genres'].apply(literal_eval)
movies_df['keywords'] = movies_df['keywords'].apply(literal_eval)
```

```python
movies_df['genres'] = movies_df['genres'].apply(lambda x : [ y['name'] for y in x])
movies_df['keywords'] = movies_df['keywords'].apply(lambda x : [ y['name'] for y in x])
movies_df[['genres', 'keywords']][:1]
```

genres'와 'keywords' 열의 문자열 데이터를 literal_eval을 사용하여 파이썬 객체로 변환
변환된 객체에서 각 장르와 키워드의 이름만 추출하여 리스트로 저장
'genres'와 'keywords' 열의 첫 번째 행을 출력

```python
from sklearn.feature_extraction.text import CountVectorizer

# CountVectorizer를 적용하기 위해 공백문자로 word 단위가 구분되는 문자열로 변환. 
movies_df['genres_literal'] = movies_df['genres'].apply(lambda x : (' ').join(x))
count_vect = CountVectorizer(min_df=0.0, ngram_range=(1,2))
genre_mat = count_vect.fit_transform(movies_df['genres_literal'])
print(genre_mat.shape)
```

'genres' 리스트를 공백으로 구분된 문자열로 변환하여 'genres_literal' 열에 저장.
CountVectorizer를 사용하여 'genres_literal' 열에서 단어 빈도 수를 기반으로 피처 매트릭스를 생성.

```python
from sklearn.metrics.pairwise import cosine_similarity

genre_sim = cosine_similarity(genre_mat, genre_mat)
print(genre_sim.shape)
print(genre_sim[:2])
```

cosine_similarity를 사용하여 장르 피처 매트릭스(genre_mat)의 코사인 유사도를 계산
계산된 유사도 행렬(genre_sim)의 형태를 출력하고 유사도 행렬의 첫 두 행을 출력하여 유사도 값을 확인

```python
genre_sim_sorted_ind = genre_sim.argsort()[:, ::-1]
print(genre_sim_sorted_ind[:1])
```

장르 유사도 행렬(genre_sim)의 각 행을 내림차순으로 정렬하여 인덱스를 반환
정렬된 인덱스를 genre_sim_sorted_ind 변수에 저장
genre_sim_sorted_ind의 첫 번째 행을 출력하여, 가장 유사한 영화들의 인덱스를 확인

```python
def find_sim_movie(df, sorted_ind, title_name, top_n=10):
    
    # 인자로 입력된 movies_df DataFrame에서 'title' 컬럼이 입력된 title_name 값인 DataFrame추출
    title_movie = df[df['title'] == title_name]
    
    # title_named을 가진 DataFrame의 index 객체를 ndarray로 반환하고 
    # sorted_ind 인자로 입력된 genre_sim_sorted_ind 객체에서 유사도 순으로 top_n 개의 index 추출
    title_index = title_movie.index.values
    similar_indexes = sorted_ind[title_index, :(top_n)]
    
    # 추출된 top_n index들 출력. top_n index는 2차원 데이터 임. 
    #dataframe에서 index로 사용하기 위해서 1차원 array로 변경
    print(similar_indexes)
    similar_indexes = similar_indexes.reshape(-1)
    
    return df.iloc[similar_indexes]
```

입력된 영화 제목(title_name)을 'title' 컬럼에서 찾아 해당 영화의 DataFrame을 추출
추출된 영화의 인덱스를 얻고, 이 인덱스를 사용하여 유사도 순으로 정렬된 인덱스 배열(sorted_ind)에서 상위 top_n개의 유사한 영화 인덱스를 추출
추출된 인덱스를 2차원 배열에서 1차원 배열로 변환
변환된 인덱스를 사용하여 유사한 영화들을 DataFrame에서 찾아 반환

```python
similar_movies = find_sim_movie(movies_df, genre_sim_sorted_ind, 'The Godfather',10)
similar_movies[['title', 'vote_average']]
```

선택한 데이터를 평균 평점(vote_average) 기준으로 내림차순 정렬하고 상위 10개의 영화를 출력.

```python
C = movies_df['vote_average'].mean()
m = movies_df['vote_count'].quantile(0.6)
print('C:',round(C,3), 'm:',round(m,3))
```

vote_average' 열의 평균을 구하여 변수 C에 저장한다.
'vote_count' 열의 60번째 백분위수(quantile)를 구하여 변수 m에 저장한다.
변수 C와 m의 값을 출력합니다. C는 전체 영화의 평균 평점을, m은 투표 수의 상위 60%에 해당하는 값임

```python
percentile = 0.6
m = movies_df['vote_count'].quantile(percentile)
C = movies_df['vote_average'].mean()

def weighted_vote_average(record):
    v = record['vote_count']
    R = record['vote_average']
    
    return ( (v/(v+m)) * R ) + ( (m/(m+v)) * C )   

movies_df['weighted_vote'] = movies_df.apply(weighted_vote_average, axis=1) 
```

```python
movies_df[['title','vote_average','weighted_vote','vote_count']].sort_values('weighted_vote',
                                                                          ascending=False)[:10]
```

percentile 변수에 0.6을 할당하여 상위 60%의 투표 수를 기준으로 설정
m 변수에는 vote_count 열의 상위 60% 백분위수 값을 할당
C 변수에는 vote_average 열의 평균 값을 할당
weighted_vote_average 함수는 각 영화의 가중 평점을 계산하는 함수로, 주어진 레코드의 투표 수(v)와 평균 평점(R)을 이용하여 계산
apply 함수를 사용하여 각 영화에 대해 weighted_vote_average 함수를 적용하여 'weighted_vote' 열을 생성
생성된 'weighted_vote' 열을 기준으로 내림차순 정렬하여 상위 10개의 영화 제목(title), 평균 평점(vote_average), 가중 평점(weighted_vote), 투표 수(vote_count)를 출력

```python
def find_sim_movie(df, sorted_ind, title_name, top_n=10):
    title_movie = df[df['title'] == title_name]
    title_index = title_movie.index.values
    
    # top_n의 2배에 해당하는 쟝르 유사성이 높은 index 추출 
    similar_indexes = sorted_ind[title_index, :(top_n*2)]
    similar_indexes = similar_indexes.reshape(-1)
# 기준 영화 index는 제외
    similar_indexes = similar_indexes[similar_indexes != title_index]
    
    # top_n의 2배에 해당하는 후보군에서 weighted_vote 높은 순으로 top_n 만큼 추출 
    return df.iloc[similar_indexes].sort_values('weighted_vote', ascending=False)[:top_n]

similar_movies = find_sim_movie(movies_df, genre_sim_sorted_ind, 'The Godfather',10)
similar_movies[['title', 'vote_average', 'weighted_vote']]
```

기준 영화의 인덱스는 제외하고 후보 영화들을 weighted_vote 열을 기준으로 내림차순 정렬하여 상위 top_n개의 영화를 선택하여 반환

변환된 평점 행렬 'ratings_matrix'의 처음 3개 행을 출력하여 확인

```python
ratings_matrix_T = ratings_matrix.transpose()
ratings_matrix_T.head(3)
```

'ratings_matrix_T' DataFrame의 처음 3개 행을 출력하여 확인합니다. 이는 영화를 행으로, 사용자를 열로 가지는 전치된 평점 행렬을 나타낸다

계산된 예측 평점 행렬을 ratings_pred_matrix라는 DataFrame으로 변환하여 사용자별로 각 영화의 예측 평점을 표시
ratings_pred_matrix의 처음 3개 사용자에 대한 예측 평점을 출력하여 확인

```python
from sklearn.metrics import mean_squared_error

# 사용자가 평점을 부여한 영화에 대해서만 예측 성능 평가 MSE 를 구함. 
def get_mse(pred, actual):
    # Ignore nonzero terms.
    pred = pred[actual.nonzero()].flatten()
    actual = actual[actual.nonzero()].flatten()
    return mean_squared_error(pred, actual)

print('아이템 기반 모든 인접 이웃 MSE: ', get_mse(ratings_pred, ratings_matrix.values ))
```

mean_squared_error 함수를 사용하여 예측된 평점(pred)과 실제 평점(actual) 간의 평균 제곱 오차(MSE)를 계산
pred와 actual 배열에서 실제 평점이 존재하는 부분만 선택하여 MSE를 계산합니다. 이는 사용자가 평점을 부여한 영화에 대해서만 평가
계산된 MSE 값을 출력하여 아이템 기반 모든 인접 이웃 방식의 예측 성능을 평가

```python
def predict_rating_topsim(ratings_arr, item_sim_arr, n=20):
    # 사용자-아이템 평점 행렬 크기만큼 0으로 채운 예측 행렬 초기화
    pred = np.zeros(ratings_arr.shape)

    # 사용자-아이템 평점 행렬의 열 크기만큼 Loop 수행. 
    for col in range(ratings_arr.shape[1]):
        # 유사도 행렬에서 유사도가 큰 순으로 n개 데이터 행렬의 index 반환
        top_n_items = [np.argsort(item_sim_arr[:, col])[:-n-1:-1]]
        # 개인화된 예측 평점을 계산
        for row in range(ratings_arr.shape[0]):
            pred[row, col] = item_sim_arr[col, :][top_n_items].dot(ratings_arr[row, :][top_n_items].T) 
            pred[row, col] /= np.sum(np.abs(item_sim_arr[col, :][top_n_items]))        
    return pred
```

ratings_arr은 사용자-아이템 평점 행렬을 나타내며, item_sim_arr은 아이템 간의 유사도 행렬을 나타냄
n은 각 아이템의 최대 유사 아이템 수를 설정하는 매개변수
pred 행렬을 초기화하고, 각 아이템에 대해 가장 유사한 상위 n개의 아이템을 기반으로 개인화된 예측 평점을 계산

ratings_matrix에서 인덱스가 9인 사용자의 평점 데이터를 선택하여 Series 형태로 가져온다
해당 사용자가 평점을 부여한 영화들 중에서 평점이 0보다 큰(평점을 부여한) 영화들을 선택
선택된 영화들을 평점이 높은 순으로 내림차순 정렬하여 상위 10개를 출력

```python
def get_unseen_movies(ratings_matrix, userId):
    # userId로 입력받은 사용자의 모든 영화정보 추출하여 Series로 반환함. 
    # 반환된 user_rating 은 영화명(title)을 index로 가지는 Series 객체임. 
    user_rating = ratings_matrix.loc[userId,:]
    
    # user_rating이 0보다 크면 기존에 관람한 영화임. 대상 index를 추출하여 list 객체로 만듬
    already_seen = user_rating[ user_rating > 0].index.tolist()
    
    # 모든 영화명을 list 객체로 만듬. 
    movies_list = ratings_matrix.columns.tolist()
    
    # list comprehension으로 already_seen에 해당하는 movie는 movies_list에서 제외함. 
    unseen_list = [ movie for movie in movies_list if movie not in already_seen]
    
    return unseen_list
```

get_unseen_movies 함수: 입력된 사용자의 평점 행렬(ratings_matrix)에서 이미 평점을 매긴 영화를 제외한 영화 리스트(unseen_list)를 반환
ratings_matrix.loc[userId, :]로 해당 사용자의 모든 영화 평점 정보를 가져온다
평점이 0보다 큰 영화는 이미 관람한 영화로 간주하고, 그 인덱스들을 already_seen 리스트로 만든다
전체 영화 리스트는 ratings_matrix.columns.tolist()로 가져오고, 리스트 컴프리헨션을 사용하여 이미 관람한 영화들을 제외한 unseen_list를 생성

```python
def recomm_movie_by_userid(pred_df, userId, unseen_list, top_n=10):
    # 예측 평점 DataFrame에서 사용자id index와 unseen_list로 들어온 영화명 컬럼을 추출하여
    # 가장 예측 평점이 높은 순으로 정렬함. 
    recomm_movies = pred_df.loc[userId, unseen_list].sort_values(ascending=False)[:top_n]
    return recomm_movies
    
# 사용자가 관람하지 않는 영화명 추출   
unseen_list = get_unseen_movies(ratings_matrix, 9)

# 아이템 기반의 인접 이웃 협업 필터링으로 영화 추천 
recomm_movies = recomm_movie_by_userid(ratings_pred_matrix, 9, unseen_list, top_n=10)

# 평점 데이타를 DataFrame으로 생성. 
recomm_movies = pd.DataFrame(data=recomm_movies.values,index=recomm_movies.index,columns=['pred_score'])
recomm_movies
```

recomm_movie_by_userid 함수: 예측 평점 행렬(pred_df)에서 입력된 사용자(userId)에 대해 unseen_list에 포함된 영화들의 예측 평점을 기준으로 상위 top_n개의 영화를 추천
pred_df.loc[userId, unseen_list]로 해당 사용자의 unseen_list에 포함된 영화들의 예측 평점을 가져온다
예측 평점을 내림차순으로 정렬하여 상위 top_n개의 영화를 선택하고 선택된 영화들과 그 예측 평점을 recomm_movies DataFrame으로 생성하여 반환

```python
import numpy as np
from sklearn.metrics import mean_squared_error

def get_rmse(R, P, Q, non_zeros):
    error = 0
    # 두개의 분해된 행렬 P와 Q.T의 내적 곱으로 예측 R 행렬 생성
    full_pred_matrix = np.dot(P, Q.T)
    
    # 실제 R 행렬에서 널이 아닌 값의 위치 인덱스 추출하여 실제 R 행렬과 예측 행렬의 RMSE 추출
    x_non_zero_ind = [non_zero[0] for non_zero in non_zeros]
    y_non_zero_ind = [non_zero[1] for non_zero in non_zeros]
    R_non_zeros = R[x_non_zero_ind, y_non_zero_ind]
    
    full_pred_matrix_non_zeros = full_pred_matrix[x_non_zero_ind, y_non_zero_ind]
      
    mse = mean_squared_error(R_non_zeros, full_pred_matrix_non_zeros)
    rmse = np.sqrt(mse)
    
    return rmse
```

get_rmse 함수: Matrix Factorization을 통해 생성된 예측 행렬과 실제 평점 행렬 간의 RMSE(Root Mean Squared Error)를 계산
입력으로 받은 R은 실제 평점 행렬을 나타낸다
P와 Q는 Matrix Factorization을 통해 얻은 사용자 요인 행렬과 아이템 요인 행렬
non_zeros는 R 행렬에서 값이 존재하는 위치의 인덱스를 나타내는 리스트
np.dot(P, Q.T)를 통해 전체 예측 행렬 full_pred_matrix를 계산.
non_zeros에서 행과 열의 인덱스를 추출하여 실제 평점 행렬(R)과 예측 행렬(full_pred_matrix)에서 값이 있는 위치의 예측 평점을 추출
추출한 실제 평점과 예측 평점을 mean_squared_error 함수를 사용하여 MSE를 계산하고, 이를 RMSE로 변환하여 반환

```python
def matrix_factorization(R, K, steps=200, learning_rate=0.01, r_lambda = 0.01):
    num_users, num_items = R.shape
    # P와 Q 매트릭스의 크기를 지정하고 정규분포를 가진 랜덤한 값으로 입력합니다. 
    np.random.seed(1)
    P = np.random.normal(scale=1./K, size=(num_users, K))
    Q = np.random.normal(scale=1./K, size=(num_items, K))
       
    # R > 0 인 행 위치, 열 위치, 값을 non_zeros 리스트 객체에 저장. 
    non_zeros = [ (i, j, R[i,j]) for i in range(num_users) for j in range(num_items) if R[i,j] > 0 ]
   
    # SGD기법으로 P와 Q 매트릭스를 계속 업데이트. 
    for step in range(steps):
        for i, j, r in non_zeros:
            # 실제 값과 예측 값의 차이인 오류 값 구함
            eij = r - np.dot(P[i, :], Q[j, :].T)
            # Regularization을 반영한 SGD 업데이트 공식 적용
            P[i,:] = P[i,:] + learning_rate*(eij * Q[j, :] - r_lambda*P[i,:])
            Q[j,:] = Q[j,:] + learning_rate*(eij * P[i, :] - r_lambda*Q[j,:])
       
        rmse = get_rmse(R, P, Q, non_zeros)
        if (step % 10) == 0 :
            print("### iteration step : ", step," rmse : ", rmse)
            
    return P, Q
```

```python
이 함수는 주어진 사용자-아이템 평점 행렬 R을 SGD(Stochastic Gradient Descent) 기법을 사용하여 잠재 요인으로 분해
각 요인을 업데이트할 때 정규화를 적용하여 과적합을 방지
매 반복마다 RMSE(Root Mean Squared Error)를 계산하여 모니터링하며, 최종적으로 업데이트된 사용자 요인 행렬 P와 아이템 요인 행렬 Q를 반환
```

```python
import pandas as pd
import numpy as np

movies = pd.read_csv('./movies.csv')
ratings = pd.read_csv('./ratings.csv')
ratings = ratings[['userId', 'movieId', 'rating']]
ratings_matrix = ratings.pivot_table('rating', index='userId', columns='movieId')

# title 컬럼을 얻기 이해 movies 와 조인 수행
rating_movies = pd.merge(ratings, movies, on='movieId')

# columns='title' 로 title 컬럼으로 pivot 수행. 
ratings_matrix = rating_movies.pivot_table('rating', index='userId', columns='title')
```

사용자-영화 평점 데이터를 사용하여 ratings_matrix라는 피벗 테이블을 생성
각 행은 사용자를 나타내고, 각 열은 영화 제목을 나타낸다.
이 테이블은 사용자가 각 영화에 매긴 평점을 포함하고 있다

```python
P, Q = matrix_factorization(ratings_matrix.values, K=50, steps=200, learning_rate=0.01, r_lambda = 0.01)
pred_matrix = np.dot(P, Q.T)
```

```python
ratings_pred_matrix = pd.DataFrame(data=pred_matrix, index= ratings_matrix.index,
                                   columns = ratings_matrix.columns)

ratings_pred_matrix.head(3)
```

주어진 사용자-영화 평점 데이터를 Matrix Factorization을 사용하여 잠재 요인 P와 Q로 분해하고, 이를 기반으로 예측 평점 행렬 ratings_pred_matrix을 생성
 각 사용자의 각 영화에 대한 예측 평점을 담고 있음

```python
def get_unseen_movies(ratings_matrix, userId):
    # userId로 입력받은 사용자의 모든 영화정보 추출하여 Series로 반환함. 
    # 반환된 user_rating 은 영화명(title)을 index로 가지는 Series 객체임. 
    user_rating = ratings_matrix.loc[userId,:]
    
    # user_rating이 0보다 크면 기존에 관람한 영화임. 대상 index를 추출하여 list 객체로 만듬
    already_seen = user_rating[ user_rating > 0].index.tolist()
    
    # 모든 영화명을 list 객체로 만듬. 
    movies_list = ratings_matrix.columns.tolist()
    
    # list comprehension으로 already_seen에 해당하는 movie는 movies_list에서 제외함. 
    unseen_list = [ movie for movie in movies_list if movie not in already_seen]
    
    return unseen_list
```

get_unseen_movies 함수:
ratings_matrix에서 주어진 userId에 해당하는 사용자의 평가 정보를 추출하여 Series로 반환
이미 평가한 영화는 already_seen에 저장하고, 이를 제외한 모든 영화명을 포함하는 movies_list를 생성
unseen_list에는 사용자가 아직 관람하지 않은 영화명들이 담겨있음

```python
def recomm_movie_by_userid(pred_df, userId, unseen_list, top_n=10):
    # 예측 평점 DataFrame에서 사용자id index와 unseen_list로 들어온 영화명 컬럼을 추출하여
    # 가장 예측 평점이 높은 순으로 정렬함. 
    recomm_movies = pred_df.loc[userId, unseen_list].sort_values(ascending=False)[:top_n]
    return recomm_movies
```

recomm_movie_by_userid 함수:
pred_df에서 userId에 해당하는 행과 unseen_list에 포함된 영화명 열을 추출하여 예측 평점이 높은 순으로 정렬
top_n 개수만큼 추천 영화를 선택하여 반환

```python
# 사용자가 관람하지 않는 영화명 추출   
unseen_list = get_unseen_movies(ratings_matrix, 9)

# 아이템 기반의 인접 이웃 협업 필터링으로 영화 추천 
recomm_movies = recomm_movie_by_userid(ratings_pred_matrix, 9, unseen_list, top_n=10)

# 평점 데이타를 DataFrame으로 생성. 
recomm_movies = pd.DataFrame(data=recomm_movies.values,index=recomm_movies.index,columns=['pred_score'])
recomm_movies
```

get_unseen_movies를 통해 사용자가 아직 관람하지 않은 영화 목록인 unseen_list를 얻는다
recomm_movie_by_userid를 사용하여 ratings_pred_matrix에서 해당 사용자에게 추천할 영화를 선정하고, 예측 평점이 높은 순으로 정렬
최종적으로 추천된 영화들을 DataFrame 형식으로 정리하여 출력
